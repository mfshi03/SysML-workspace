```sysml
package LandmineDetectionSystem {

    import Base::*, Interactions::*, Activities::*, StateMachines::*, Allocations::*;
    
    part def BaseSecurityControlCenter {
        ref link: CommunicationLink;
        action sendCommands;
        action receiveData;
    }

    part def UAS {
        ref link: CommunicationLink;
        action detectLandmineUsingAI;
        action sendDataToBase;
        action receiveCommandsFromBase;
    }

    part def UGV {
        ref link: CommunicationLink;
        action detectLandmineUsingOpticSensor;
        action sendDataToBase;
        action receiveCommandsFromBase;
    }

    block def CommunicationLink {
        port out operationalVideo: DataStream;
        port out reconnaissanceData: DataStream;
        port out vehicleDiagnosticData: DataStream;
        port in vehicleCommands: CommandStream;
    }

    block def DataStream {
        attribute dataContent: String;
    }

    block def CommandStream {
        attribute commandContent: String;
    }

    block def Terrain {
        attribute terrainType: String;
    }

    item flow def VideoStream extends DataStream {
        attribute videoContent: String;
    }

    item flow def ReconnaissanceStream extends DataStream {
        attribute reconContent: String;
    }

    item flow def DiagnosticStream extends DataStream {
        attribute diagnosticContent: String;
    }

    item flow def CommandStream {
        attribute commandContent: String;
    }

    connection linkConnectionA between UAS::link and UGV::link {
        item flow video: VideoStream;
        item flow recon: ReconnaissanceStream;
        item flow diagnostic: DiagnosticStream;
        item flow commands: CommandStream;
    }

    connection linkConnectionB between BaseSecurityControlCenter::link and UAS::link {
        item flow commands: CommandStream;
        item flow video: VideoStream;
        item flow recon: ReconnaissanceStream;
        item flow diagnostic: DiagnosticStream;
    }

    connection linkConnectionC between BaseSecurityControlCenter::link and UGV::link {
        item flow commands: CommandStream;
        item flow recon: ReconnaissanceStream;
        item flow diagnostic: DiagnosticStream;
    }

}
```
